```python
#Importing the required libraries
%matplotlib inline
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import RandomizedSearchCV,cross_val_predict
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
```

```python
#Reading the training data set
train = pd.read_csv('train.csv')
train.info()
```

```python
train.describe()   #Data is clean with no null entries
```

```python
train.head()   
```

```python
train_label = train['label']    #Separating Label data for target set
train = train.drop(['label'],axis=1)    #Dropping the target data set
```

```python
random_digit = np.array(train.loc[30000,:])    #Viewing one of the data 
type(random_digit)
```

```python
random_digit = random_digit.reshape(28,28)             #Reshaping to 28*28 matrix
plt.imshow(random_digit,cmap = matplotlib.cm.binary)   #Plotting the data
plt.axis('off')
plt.show()
```

```python
train = train.astype('float32')    #Converting to float for normalizing the data
train = train/255                  #Normalizing to reduce variance
```

```python
#Building the model
model = RandomForestClassifier()   #Building the Classifier
tuning_params = {'n_estimators':np.arange(1,15)}   #HyperTuning the model
model_tuning = RandomizedSearchCV(model,tuning_params,cv=5)  #Using RandomizedSearchCV for faster computation
model_tuning.fit(train,train_label)        #Fine Tuning the model
print(model_tuning.best_params_)           #Printing the best param
```

```python
updated_model = RandomForestClassifier(n_estimators=14)    #Rebuilding model with the best param
updated_model.fit(train,train_label)                       #Training the model
test = pd.read_csv('test.csv')                             #Reading the test data
test.head()
```

```python
test = test.astype('float32')    #Normalizing as in training data
test = test/255
```

```python
predict = updated_model.predict(test)    #Predicting the values
```

```python
#Testing the accuracy of the model/Estimatin the model
train_proba = cross_val_predict(updated_model,train,train_label,cv=3)  
cnf = confusion_matrix(train_label,train_proba)
plt.matshow(cnf,cmap=plt.cm.gray)
plt.show()    #Bulding a confusion matrix to check where the model has gone wrong
```

```python
updated_model.score(train,train_label)   #Accuracy is great on training data
```

```python
ImageId = np.arange(1,28001)        
```

```python
final_predict = pd.DataFrame({'ImageId': ImageId, 'Label': predict}, columns=['ImageId', 'Label'])    #Bulding the final DataFrame
final_predict.head()   #Data looks clean
```

```python
final_predict.to_csv('final_predict.csv',index=False)    #parsing data to a csv file
```
